```python
from cigre_eu_mv import cigre_eu_mv_class
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sopt
import ipywidgets
from pydae import ssa
```

```python
syst = cigre_eu_mv_class()
syst.Dt = 10000e-6
syst.decimation = 1
syst.N_store = 100_000
syst.update()
```

```python
syst.sopt_root_jac = False
events = [{'t_end':0.01}]

syst.initialize(events,xy0=10e3)
```

```python
for item in syst.x_list:
    print(f'{item:8s} = {syst.get_value(item):8.2f}')
```

```python
events = [{'t_end':0.001,  'i_11_Q':0.0},
          {'t_end':0.1,  'i_11_Q':100.0},]
syst.simulate(events,xy0=20e3);
```

```python
plt.close('all')
fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(10, 5), frameon=False, dpi=80)

axes[0,0].plot(syst.T, syst.get_values('v_10_D'), label=f"Bus 10: $v_D$")
axes[0,0].plot(syst.T, syst.get_values('v_10_Q'), label=f"Bus 10: $v_Q$")
axes[0,1].plot(syst.T, syst.get_values('i_l_1011_D'), label=f"Line 10-11: $i_D$")
axes[0,1].plot(syst.T, syst.get_values('i_l_1011_Q'), label=f"Line 10-11: $i_Q$")
#axes[1,1].plot(times,X_obs[:,0] + x_d_0[0],   label=f'$X_obs$')
for ax in axes.flatten():
    ax.legend(loc='best')
    ax.grid(True)
```
