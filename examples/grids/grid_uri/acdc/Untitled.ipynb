```python
#alpha = np.exp(2.0/3*np.pi*1j)
a_r,a_i = sym.symbols('a_r,a_i',real=True)
alpha = a_r + sym.I*a_i
A_0a =  sym.Matrix([[1, 1, 1],
                  [1, alpha**2, alpha],
                  [1, alpha, alpha**2]])

A_a0 = 1/3* sym.Matrix([[1, 1, 1],
                                     [1, alpha, alpha**2],
                                     [1, alpha**2, alpha]])
```

```python
v_a_r,v_b_r,v_c_r = sym.symbols('v_a_r,v_b_r,v_c_r',real=True)
v_a_i,v_b_i,v_c_i = sym.symbols('v_a_i,v_b_i,v_c_i',real=True)
v_abc = sym.Matrix([[v_a_r+sym.I*v_a_i],[v_b_r+sym.I*v_b_i],[v_c_r+sym.I*v_c_i]])
i_a_r,i_b_r,i_c_r = sym.symbols('i_a_r,i_b_r,i_c_r',real=True)
i_a_i,i_b_i,i_c_i = sym.symbols('i_a_i,i_b_i,i_c_i',real=True)
i_abc = sym.Matrix([[i_a_r+sym.I*i_a_i],[i_b_r+sym.I*i_b_i],[i_c_r+sym.I*i_c_i]])
```

```python
v_opn = A_a0*v_abc
i_opn = A_a0*i_abc
```

```python
v_0 = v_opn[0]
v_p = v_opn[1]
v_n = v_opn[2]
i_0 = i_opn[0]
i_p = i_opn[1]
i_n = i_opn[2]
```

```python
s_abc = sym.conjugate(i_abc.T)*(v_abc)
s_opn = 3*sym.conjugate(i_opn.T)*(v_opn)
```
