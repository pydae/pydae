## Formulation

### Grid

$$
\left[ {\begin{array}{cc}
   \mathbf Y_{vv} &  \mathbf Y_{vi} \\
   \mathbf Y_{iv} &  \mathbf Y_{ii} \\
\end{array} } \right]
\left[ {\begin{array}{c}
   \mathbf{\underline  v}_k \\
   \mathbf{\underline  v}_u \\
\end{array} } \right]
=
\left[ {\begin{array}{c}
   \mathbf{\underline  i}_u \\
   \mathbf{\underline  i}_k \\
\end{array} } \right]
$$

$$
\mathbf g_{cplx} = -\mathbf Y_{ii}  \mathbf{\underline  v}_u  - \mathbf Y_{iv} \mathbf{\underline  v}_k
$$

### Loads

Phasor voltages and neutral voltage are defined as:

$$
    \underline v_k = v_{lk}^r + jv_{lk}^i  \;\;\;\;\;\;\;\;  k = a,\,b,\,c,\,n
$$    
  
  
Phasor currents and neutral currents are defined as:

$$
    \underline i_k = i_{lk}^r + ji_{lk}^i  \;\;\;\;\;\;\;\;  k = a,\,b,\,c,\,n
$$

The vectors for the new algebraic equations $\mathbf g_l$, the new algebraic states $\mathbf y_l$ and the new inputs $\mathbf u_l$ are defined as follows:

$$
    \mathbf g_l =
    \left[
    \begin{array}{c}
     p_a + \Re \left(\left(\underline v_a - \underline v_n\right) \underline i_a^*\right) \\ 
     p_b + \Re \left(\left(\underline v_b - \underline v_n\right) \underline i_b^*\right) \\
     p_c + \Re \left(\left(\underline v_c - \underline v_n\right) \underline i_c^*\right) \\
     q_a + \Im \left(\left(\underline v_a - \underline v_n\right) \underline i_a^*\right) \\
     q_b + \Im \left(\left(\underline v_b - \underline v_n\right) \underline i_b^*\right) \\
     q_c + \Im \left(\left(\underline v_c - \underline v_n\right) \underline i_c^*\right) \\
     \Re \left(\underline i_a+ \underline i_b+ \underline i_c+ \underline i_n\right) \\
     \Im \left(\underline i_a+ \underline i_b+ \underline i_c+ \underline i_n\right) \\
    \end{array} 
    \right]
    \;\;\;\;\;\;\;
    \mathbf y_l =
    \left[
    \begin{array}{c}
     i_{la}^r \\ 
     i_{lb}^r \\
     i_{lc}^r \\
     i_{la}^i \\
     i_{lb}^i \\
     i_{lc}^i \\
     i_{ln}^r \\
     i_{ln}^i \\
    \end{array} 
    \right]
        \;\;\;\;\;\;\;
        \mathbf u_l =
    \left[
    \begin{array}{c}
     p_a \\ 
     p_b  \\
     p_c  \\
     q_a  \\
     q_b  \\
     q_c  \\
    \end{array} 
    \right]
$$



### VSC feeder

#### AC side

$$
    \underline v_k = v_{lk}^r + jv_{lk}^i  \;\;\;\;\;\;\;\;  k = a,\,b,\,c,\,n
$$    
  
  
Phasor currents and neutral currents are defined as:

$$
    \underline i_k = i_{lk}^r + ji_{lk}^i  \;\;\;\;\;\;\;\;  k = a,\,b,\,c,\,n
$$

The vectors for the new algebraic equations $\mathbf g_l$, the new algebraic states $\mathbf y_l$ and the new inputs $\mathbf u_l$ are defined as follows:

$$
    \mathbf g_{vsc} =
    \left[
    \begin{array}{c}
     p_a + \Re \left(\left(\underline v_a - \underline v_n\right) \underline i_a^*\right) \\ 
     p_b + \Re \left(\left(\underline v_b - \underline v_n\right) \underline i_b^*\right) \\
     p_c + \Re \left(\left(\underline v_c - \underline v_n\right) \underline i_c^*\right) \\
     q_a + \Im \left(\left(\underline v_a - \underline v_n\right) \underline i_a^*\right) \\
     q_b + \Im \left(\left(\underline v_b - \underline v_n\right) \underline i_b^*\right) \\
     q_c + \Im \left(\left(\underline v_c - \underline v_n\right) \underline i_c^*\right) \\
     \Re \left(\underline i_a+ \underline i_b+ \underline i_c+ \underline i_n\right) \\
     \Im \left(\underline i_a+ \underline i_b+ \underline i_c+ \underline i_n\right) \\
    \end{array} 
    \right]
    \;\;\;\;\;\;\;
    \mathbf y_{vsc} =
    \left[
    \begin{array}{c}
     i_{la}^r \\ 
     i_{lb}^r \\
     i_{lc}^r \\
     i_{la}^i \\
     i_{lb}^i \\
     i_{lc}^i \\
     i_{ln}^r \\
     i_{ln}^i \\
    \end{array} 
    \right]
        \;\;\;\;\;\;\;
        \mathbf u_{vsc} =
    \left[
    \begin{array}{c}
     p_a \\ 
     p_b  \\
     p_c  \\
     q_a  \\
     q_b  \\
     q_c  \\
    \end{array} 
    \right]
$$

#### Losses

$$ i_{k}^{rms} = \sqrt{ i_{kr}^2 + i_{ki}^2 } \;\;\;\;\;\;\;\;  k = a,\,b,\,c,\,n $$ 

$$ p_{loss}^k  = A + B i_{k}^{rms} + C \left( i_{k}^{rms} \right)^2   \;\;\;\;\;\;\;\;  k = a,\,b,\,c,\,n $$          

$$ p_{loss} = p_{loss}^a + p_{loss}^b +p_{loss}^c + p_{loss}^n$$

#### DC side

$$ p_{dc} = p_{ac} - p_{loss} $$

$$ i_{dc} = \frac{p_{dc}}{v_{dc}} $$

$$ v_{dc} = v_{ar} - v_{nr} $$ 

$$ i_{ar} = i_{dc} $$
$$ i_{nr} = - i_{dc}$$


#### Sintax

##### Inputs

* $p_a$,  $p_b$,  $p_b$:  `p_vsc_a_<ac bus name>`, `p_vsc_b_<ac bus name>`, `p_vsc_c_<ac bus name>`
* $q_a$,  $q_b$,  $q_b$:  `q_vsc_a_<ac bus name>`, `q_vsc_b_<ac bus name>`, `q_vsc_c_<ac bus name>`

##### Some algebraic states 

* $p_{dc}$:  `p_vsc_<dc bus name>`
* $p_{loss}$:  `p_vsc_loss_<ac bus name>`
