```python
import numpy as np
import matplotlib.pyplot as plt
import temp

from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
model=temp.model()

p_load_A2 = 50.0e3
q_load_A2 = 0.0
params = {'p_c_A1':0.0, 'K_delta_A1':0.1, 'K_qp_A1':0.0, 'K_qi_A1':0.0,
"e_ao_m_A1": 230.94*1.003,
"e_bo_m_A1": 230.94*1.003,
"e_co_m_A1": 230.94*1.003,
"p_load_A2_a" :p_load_A2/3,
"q_load_A2_a" :q_load_A2/3,
"p_load_A2_b" :p_load_A2/3,
"q_load_A2_b" :q_load_A2/3,
"p_load_A2_c" :p_load_A2/3,
"q_load_A2_c" :q_load_A2/3,
'i_dc_ref_A1':0.0, 'U_opt_A1':0.7
}  


model.ini(params,'xy_0.json')
```

```python
model = temp.model()
model.Dt = 0.01
model.ini(params,xy_0='xy_0.json')
model.ini({'p_c_A1':0.0,'K_delta_A1':0.0})
model.run(1,{})

p_load_A2 = 100.0e3
model.run(40,{"p_load_A2_a" :p_load_A2/3,
              "q_load_A2_a" :q_load_A2/3,
              "p_load_A2_b" :p_load_A2/3,
              "q_load_A2_b" :q_load_A2/3,
              "p_load_A2_c" :p_load_A2/3,
              "q_load_A2_c" :q_load_A2/3}  )
model.post();
```

```python
fig = make_subplots(rows=2, cols=1)

fig.add_trace(go.Scatter(x=model.Time, y=model.get_values('p_pos_A1'), mode='lines', name='p_pos_A1 '), row=1, col=1)
fig.add_trace(go.Scatter(x=model.Time, y=model.get_values('q_pos_A1'), mode='lines', name='q_pos_A1 '), row=1, col=1)

fig.add_trace(go.Scatter(x=model.Time, y=model.get_values('omega_A1'), mode='lines', name='omega_A1 '), row=2, col=1)

fig.update_layout(height=600, width=800, title_text="Side By Side Subplots")
fig.show()
```

```python
fig = make_subplots(rows=2, cols=2)

fig.add_trace(go.Scatter(x=model.Time, y=model.get_values('v_fc_dc_A1'), mode='lines', name='v_fc_dc_A1'), row=1, col=1)
fig.add_trace(go.Scatter(x=model.Time, y=model.get_values('v_dc_A1'), mode='lines', name='v_dc_A1'), row=1, col=1)
fig.add_trace(go.Scatter(x=model.Time, y=model.get_values('q_h2_A1'), mode='lines', name='q_h2_A1 '), row=2, col=1)
fig.add_trace(go.Scatter(x=model.Time, y=model.get_values('p_h2_A1'), mode='lines', name='p_h2_A1 '), row=1, col=2)

fig.update_layout(height=600, width=800, title_text="Side By Side Subplots")
fig.show()
```

```python
model.report_u()
model.report_x()
model.report_y()
model.report_z()
```
