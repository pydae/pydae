```python
model=temp.model()
model.Dt = 0.1
model.ini({'p_load_D2':430*50, 'i_dc_ref_D1':0.0, 'U_opt_D1':0.7},'xy_0.json')
model.run(50.0,{})
# model.run(20,{'p_load_D2':0.001e3})
# model.run(40,{'p_load_D2':0.002e3})
# model.run(200,{'p_load_D2':0.005e3})
model.run(200,{'p_load_D2':430*20})
model.run(400,{'p_load_D2':430*10})

model.run(800,{'p_load_D2':430*100})

model.post();
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(model.Time,model.get_values('v_dc_D1'), label='v_dc_D1')
axes[1,0].plot(model.Time,model.get_values('i_dc_D1'), label='i_dc_D1')

axes[0,1].plot(model.Time,  model.get_values('p_h2_D1'),label='p_h2')
axes[0,1].plot(model.Time,  model.get_values('p_h2o_D1'),label='p_h2o')
axes[0,1].plot(model.Time,  model.get_values('p_o2_D1'),label='p_o2')

axes[1,1].plot(model.Time,  model.get_values('q_h2_D1'),label='q_h2_D1')

for ax in axes.flatten():
    ax.grid()
    ax.legend()
    ax.set_xlim(0,model.Time[-1])
axes[1,0].set_xlabel('Time (s)')
axes[1,1].set_xlabel('Time (s)')
```
