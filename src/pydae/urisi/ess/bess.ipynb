```python
import numpy as np
import temp
import matplotlib.pyplot as plt

S_n = 100e3
V_n = 800
I_n = S_n/V_n
Conduction_losses = 0.02*S_n # = A*I_n**2
losses = 1.0
A = Conduction_losses/(I_n**2)*losses
B = 1*losses
C = 0.02*S_n*losses
model = temp.model()
model.Dt = 1.0
params = {'A_loss_A1':A,'B_loss_A1':B,'C_loss_A1':C,'soc_ref_A1':1.0,'R_bat_A1':0.1,
          'K_soc_p_A1':0.0,'K_soc_i_A1':1e-6}
model.ini(params,'xy_1.json')
# model.save_xy_0('xy_1.json')
# model.report_params()
# model.report_u()
model.report_x()
model.report_y()
# model.report_z()
```

```python
model = temp.model()
model.Dt = 1.0
model.ini(params,'xy_0.json')
model.run(3600,{'p_ref_A1':90e3})
model.run(7200,{'p_ref_A1':-90e3})

model.post();
model.report_x()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(model.Time,model.get_values('soc_A1'), label='soc')
axes[0,1].plot(model.Time,model.get_values('p_A2'), label='Grid p')
axes[1,0].plot(model.Time,model.get_values('v_dc_A1'), label='v_dc')
for ax in axes.flatten():
    ax.legend()
    ax.grid()
```
