```python
H = 100
P = 100e6
rend = 0.9
rho = 1000
g = 9.8

Q = P/(rend*rho*g*H)
print(f'Q = {Q:0.2f} m³/s')
```

```python
D=0.835*Q**(2/5)
S=np.pi*D*(2/4)

V = Q/S

L = 200 #m

g = 9.81 # m/s^2

T_w = L*Q/(g*S*H)

print(f'T_w = {T_w:0.2f} s')
```

```python
model = temp.model()
model.Dt = 0.01
model.ini({'P_2':-80e6,'T_r_gov_1':5,'R_r_gov_1':1.5,'T_w_gov_1':T_w,
           'H_1':2},'xy_0.json')
model.run(1.0,{})
model.run(60.0,{'P_2':-50e6})
model.post()

fig,axes = plt.subplots(nrows=4, figsize=(4,7))
axes[0].plot(model.Time,model.get_values('p_m_1'), label='$p_m$')
axes[1].plot(model.Time,model.get_values('omega_coi'), label='$\omega_{coi}$')
axes[2].plot(model.Time,model.get_values('g_gov_1'), label='$g$')
axes[2].plot(model.Time,model.get_values('q_gov_1'), label='$q$')
axes[3].plot(model.Time,model.get_values('h_gov_1'), label='$h$')

for ax in axes:
    ax.legend()
    ax.grid()
fig.savefig('hygov_step.svg')
```
